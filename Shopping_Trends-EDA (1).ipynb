```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
shop = pd.read_csv('D:/Jupyter Projects/shopping_trends.csv')
shop['Age_category'] = pd.cut(shop['Age'], bins=[0, 15, 18, 30, 50, 70], labels=['Child', 'Teen', 'Young Adult', 'Middle-Aged Adults', 'Old'])
age_avg = shop.groupby('Age_category')['Age'].mean().reset_index()
age_avg = shop.groupby('Age_category', as_index=False)['Age'].mean()
fig = px.bar(age_avg, x='Age_category', y='Age', 
             title='Average Age by Category',
             labels={'Age': 'Average Age', 'Age_category': 'Age Category'})
fig.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
shop=pd.read_csv('D:/Jupyter Projects/shopping_trends.csv')
shop.head()
```

```python
age_gender_trends = shop.groupby(['Age', 'Gender']).agg({
    'Purchase Amount (USD)': ['mean', 'sum', 'max', 'min', 'std'], 
    'Category': lambda x: x.mode()[0],
    'Frequency of Purchases': lambda x: x.mode()[0]
}).reset_index()

age_gender_trends.columns = [' '.join(col).strip() for col in age_gender_trends.columns.values]

age_gender_trends = age_gender_trends.head(10)

age_gender_trends['Row Number'] = range(1, len(age_gender_trends) + 1)

age_gender_trends = age_gender_trends[['Row Number'] + [col for col in age_gender_trends.columns if col != 'Row Number']]

table = age_gender_trends.style.set_table_styles(
    [{
        'selector': 'thead th',
        'props': [('font-weight', 'bold')]
    }, {
        'selector': 'table',
        'props': [('border', '1px solid black'), ('border-collapse', 'collapse')] 
    }, {
        'selector': 'td, th',
        'props': [('border', '1px solid black')]  
    }]
)
table
```

```python
loyalty_data = shop.groupby('Subscription Status').agg({
    'Purchase Amount (USD)': 'mean',
    'Previous Purchases': 'mean'
}).reset_index()

styled_table = loyalty_data.style.set_table_styles(
    [{
        'selector': 'thead th',
        'props': [('font-weight', 'bold')]
    }, {
        'selector': 'table',
        'props': [('border', '1px solid black'), ('border-collapse', 'collapse')] 
    }, {
        'selector': 'td, th',
        'props': [('border', '1px solid black')] 
    }]
)
styled_table
```

```python
payment_shipping = shop.groupby('Payment Method').agg({
    'Payment Method': 'count',
    'Shipping Type': lambda x: x.mode()[0]
}).rename(columns={'Payment Method': 'Count'}).reset_index()

table = payment_shipping.style.set_table_styles(
    [{
        'selector': 'thead th',
        'props': [('font-weight', 'bold')]
    }, {
        'selector': 'table',
        'props': [('border', '1px solid black'), ('border-collapse', 'collapse')] 
    }, {
        'selector': 'td, th',
        'props': [('border', '1px solid black')] 
    }]
)
table
```

```python
discount_ratings = shop.groupby('Discount Applied').agg({
    'Review Rating': ['mean', 'max', 'min', 'std'], 
    'Purchase Amount (USD)': 'count' 
}).rename(columns={'Purchase Amount (USD)': 'Purchase Volume'}).reset_index()

discount_ratings.columns = [' '.join(col).strip() for col in discount_ratings.columns.values]

table = discount_ratings.style.set_table_styles(
    [{
        'selector': 'thead th',
        'props': [('font-weight', 'bold')] 
    }, {
        'selector': 'table',
        'props': [('border', '1px solid black'), ('border-collapse', 'collapse')] 
    }, {
        'selector': 'td, th',
        'props': [('border', '1px solid black')]
    }]
)
table
```
